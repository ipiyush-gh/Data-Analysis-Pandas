```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\Admin\OneDrive\Desktop\BCA\TY,BCA\InternSpark Internship\Task1-Data Analysis Project (Pandas)\sales_data.csv')

print("First 5 rows:")
print(df.head())

print("\nMissing Values:")
print(df.isnull().sum())

df = df.dropna()
df = df.drop_duplicates()

print("\nSummary Statistics:")
print(df.describe())

category_sales = df.groupby('Category')['Sales'].sum()
print("\nTotal Sales by Category:")
print(category_sales)

city_sales = df.groupby('City')['Sales'].sum()
print("\nTotal Sales by City:")
print(city_sales)

high_sales = df[df['Sales'] > 500]
print("\nHigh Sales Data:")
print(high_sales)

category_sales.plot(kind='bar', title='Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```
